```python
#open the txt file from https://www.worldjigsawpuzzle.org/
file= open("finale_coppia.txt","r", encoding='utf-8')
```

```python
#open it in a string and remove the indentation
stringa = file.read()
stringa=stringa.replace('\n','')
stringa
```

```python
import pandas as pd
import numpy as np
import re
from datetime import datetime
```

```python
#create a dataframe
#posizione is the ranking the contestant placed in
#nazione is thei nation
#nome is thei full name
#tempo is the total time in which they finished the puzzle
#scarto1 is the time between each contestant and the first place
#scarto2 is the time between each contestant and the one right before them
df = pd.DataFrame(columns=["Posizione", "Nazione", "Nome","Tempo","Scarto1","Scarto2"])
```

```python
#split the string in an array so that per each line we have the information about a single contestant
vet=stringa.split('</tr>')
vet
```

```python
#take the ranking number using the indexes of unique elements of the strings and put then in a list
#we take the positions instead of using the indexes that the dataframe will produce because there is the possibility of contestant having the same time and so the same position
pos=[]
for str in vet:
    ind1=str.find("puesto\">")
    ind2=str.find('<',ind1)
    pos.append(str[ind1+8:ind2])
pos
```

```python
#same thing but for the nation
naz=[]
for str in vet:
    ind0=str.find(" title=\"")
    ind1=str.find(" title=\"", ind0+10)
    ind2=str.find('\"',ind1+8)
    naz.append(str[ind1+8:ind2])
naz
```

```python
#same thing but for the name
nome=[]
for str in vet:
    ind1=str.find("</div><div>")
    ind2=str.find('</div>\t\t<div class="ver_movil"',ind1)
    nome.append(str[ind1+11:ind2-4].replace(' <br>',',' )) #for the couple competitions the names of the 2 contestants are divided by <br>
nome
```

```python
#same thing but for scarto2
#for the contestants who didn't finish it won't find the strings because the times weren't recorded, so the indexes will be -1. in this case we put NaN
#the same happen for the first place
sca2=[]
for str in vet:
    ind1=str.find("+")
    ind2=str.find('(',ind1)
    if (ind1-ind2)==0:
        sca2.append('NaN')
    else:    
        sca2.append(str[ind1+1:ind2-1])
sca2
```

```python
#same thing as name and nation but for the total time
temp=[]
for str in vet:
    ind1=str.find("tiempo\">")
    ind2=str.find(' <',ind1)
 
    temp.append(str[ind1+8:ind2])
temp
```

```python
#same thing as scarto2
sca=[]
for str in vet:
    ind1=str.find("( alle")
    ind2=str.find('<',ind1)
    if (ind1-ind2)==0:
        sca.append('NaN')
    else:    
        sca.append(str[ind1+8:ind2-1])

sca
```

```python
#put the informations in the dataframe
df['Posizione']=pos
df['Nome']=nome
df['Nazione']=naz
df['Scarto1']=sca
df['Tempo']=temp
df['Scarto2']=sca2
```

```python
#delete the last row because it's empty due to how we initially split the string
df=df.iloc[:-1,:]
df
```

```python
#make a function to convert the time in the right format. for the contestants that didn't finish in time they reported the number of pieces left
#we don't need it so we'll only report that they didn't finish in time
def converti_tempo(tempo_tot):
    try:
        return datetime.strptime(tempo_tot, "%H:%M:%S").time()
    except ValueError:
       return 'Non ha finito in tempo'  #output if the format is not recognised


df["tempo_time"] = df["Tempo"].apply(converti_tempo)
df["tempo_time"]
```

```python
#make a function to convert the time differences in the right format.
def converti_orario(orario):
    formati = ["%M\' %S\'\'"," %S\'\'","%M\' %S\'"," %S\'"]
    for formato in formati:
        try:
            return datetime.strptime(orario, formato).time()
        except ValueError:
            continue
    return 'NaN'  #output if the format is not recognised

# apply the function for scarto1 and scarto2
df["scarto1_time"] = df["Scarto1"].apply(converti_orario)

df["scarto2_time"] = df["Scarto2"].apply(converti_orario)
```

```python
#control to make sure that the only time that have a NaN value are the first contestant and tha last ones that didn't finish.
#if there is a NaN value in the middle of the dataframe it means there i a time format that wasn't recognised correctly
df.loc[df['scarto1_time']=='NaN']
```

```python
#same control for the value scarto2_time
df.loc[df['scarto2_time']=='NaN']
```

```python
#keep only the relevant and formatted columns

df_finale=df[['Posizione','Nazione','Nome','tempo_time','scarto1_time','scarto2_time']]
df_finale
```
